```python
#popularity based recommender system
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Check the data types of all columns in the DataFrame
print(ratings_with_name.dtypes)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
rated = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(rated)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = y[y].index
```

```python
final_ratings =filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
  # index fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

  data = []
  for i in similar_items:
    item = []
    temp_df = books[books['Book-Title'] == pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    data.append(item)
  return data
```
